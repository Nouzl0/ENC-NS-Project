```python
import sys
from pathlib import Path

sys.path.insert(0, "../src")

import torch
from train import Detector, _DETECT_TRANSFORM
from canvas import DigitCanvas
from models import ModelFactory

# Load checkpoint and fix the stored model_name if it includes the norm suffix
_KNOWN_NORMS = {"none", "minmax", "standardize", "symmetric", "augmented"}

ckpt_path = "MiniResNet_augmented.pt"
data = torch.load(ckpt_path, map_location="cpu", weights_only=True)

stored_name = data["model_name"]
arch_name = next(
    (stored_name.removesuffix(f"_{n}") for n in _KNOWN_NORMS if stored_name.endswith(f"_{n}")),
    stored_name,
)
data["model_name"] = arch_name
torch.save(data, ckpt_path)  # patch the file once
print(f"Patched model_name: {stored_name!r} → {arch_name!r}")

detector = Detector(ckpt_path)
print("Model loaded.")

DigitCanvas(detector, width=280, height=280, radius=4).show()
```
